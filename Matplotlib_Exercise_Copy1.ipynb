```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

# if we are not using jupyter notebook, that line above is not required. 
# Just use plt.show() to display figures
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y, 'y')
plt.xlabel('Usia')
plt.ylabel('Gaji')
plt.title('Grafik Usia Terhadap Gaji')
plt.show()
```

```python
plt.subplot(2, 3, 1)
plt.plot(x, y, 'r--')
plt.subplot(2, 3, 2)
plt.plot(y, x, 'g*-')
plt.subplot(2, 3, 3)
plt.plot(x, y, 'b--')
plt.subplot(2, 3, 4)
plt.plot(y, x, 'y*-')
plt.subplot(2, 3, 5)
plt.plot(x, y, 'c--')
plt.subplot(2, 3, 6)
plt.plot(y, x, '*-')
plt.show()
```

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y, 'b')
axes.set_xlabel('Set X Label')
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.3, 0.4, 0.3])

axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes1')
axes1.set_ylabel('Y_label_axes1')
axes1.set_title('Axes 1 Title')

axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes1')
axes2.set_ylabel('Y_label_axes1')
axes2.set_title('Axes 2 Title')
```

```python
fig, axes = plt.subplots()

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1)
```

```python
for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12,7), dpi=100)

ax1 = axes[0][2]
np.random.seed(7)
data = [np.random.normal(0, std, 100) for std in range(1,4)]
ax1.boxplot(data, vert=False, patch_artist=True, whis=5)

from random import sample
data = sample(range(1, 1000), 100)
ax2 = axes[1][1]
ax2.hist(data)

ax3 = axes[1][0]
ax3.plot(x, x**3, 'g', label='x**3')
ax3.plot(x, y, 'r--', label='x**2')
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.set_title('title')
ax3.legend()

ax4 = axes[1][2]
ax4.plot(x, y, 'm*', label='expense')
ax4.set_xlabel('holidays')
ax4.set_ylabel('expenses')
ax4.set_title('holidays vs expense')
ax4.legend()

ax5 = axes[0][0]
ax5.plot(x, -y, color='#DEA1A1', lw= 10, ls=':', marker='+', markersize=40, alpha=75, label='performance')
ax5.set_xlabel('age')
ax5.set_ylabel('performance')
ax5.set_title('age vs performance')
ax5.set_xlim(0, 3)
ax5.set_ylim(-100, 0)
ax5.legend(loc=0)

ax6 = axes[0][1]
ax6.scatter(x,y)

plt.tight_layout()
```
